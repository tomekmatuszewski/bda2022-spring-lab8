```
from pyspark.ml.linalg import Vectors
from pyspark.ml.feature import VectorAssembler
```

```
dataset = spark.createDataFrame(
    [(1, 680, 1994, Vectors.dense([4, 1, 1]), 654900),
     (2, 840, 1994, Vectors.dense([5, 2, 1]), 898400),
     (3, 567, 1994, Vectors.dense([3, 1, 1]), 548700)],
    ["id", "size", "year", "rooms_bedrooms_kitchens", "price"])
```

```
assembler = VectorAssembler(
    inputCols=["size", "year", "rooms_bedrooms_kitchens"],
    outputCol="features")
```

```
output = assembler.transform(dataset)
print("Assembled columns 'size', 'year', 'rooms_bedrooms_kitchens' to vector column 'features'")
output.select("features", "price").show(truncate=False)
```
